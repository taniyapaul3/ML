```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("canada_per_capita_income.csv")

# Display first few rows
print(df.head())

# Define independent and dependent variables
X = df[['year']]          # Independent variable must be 2D
y = df['per capita income (US$)']

# Create linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict income for year 2020
year_2020 = np.array([[2020]])
predicted_income = model.predict(year_2020)

print("Predicted per capita income for Canada in 2020: ", predicted_income[0])

# Optional: Plot regression line
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Canada Per Capita Income Prediction")
plt.show()
```

```python
# Import required libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("hiring.csv")

# Display dataset
print(df)

# -------------------------------
# DATA PREPROCESSING
# -------------------------------

# Replace word-based experience with numbers
df['experience'] = df['experience'].replace({
    'zero': 0,
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9,
    'ten': 10,
    'eleven': 11,
    'twelve': 12
})

# Replace missing experience with 0
df['experience'] = df['experience'].fillna(0)

# Replace missing test scores with median
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(
    df['test_score(out of 10)'].median()
)

# -------------------------------
# MODEL BUILDING
# -------------------------------

X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

model = LinearRegression()
model.fit(X, y)

# -------------------------------
# PREDICTIONS
# -------------------------------

salary_1 = model.predict([[2, 9, 6]])
salary_2 = model.predict([[12, 10, 10]])

print("Predicted salary for 2 yr exp, 9 test, 6 interview:", salary_1[0])
print("Predicted salary for 12 yr exp, 10 test, 10 interview:", salary_2[0])
```

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("salary.csv")

# Check for missing values
print(df.isnull().sum())

# Remove rows with missing values
df = df.dropna()

# Define X and y
X = df[['YearsExperience']]   # must be 2D
y = df['Salary']

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Predict salary for 12 years experience
predicted_salary = model.predict([[12]])

print("Predicted salary for 12 years experience:", predicted_salary[0])

# Plot regression line
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary Prediction using Linear Regression")
plt.show()
```
